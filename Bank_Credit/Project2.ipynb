###  0.9  Ответы на вопросы.

##### 0.9.0.1 Вопрос 1:

Судя по таблице № 3, составленной по методу .pivot_table(), имеется зависимость между количеством детей и возвратом кредита в срок.

Формулируется однозначным утверждением: чем меньше количество детей у данных клиентов банка, тем выше вероятность возрата кредита в срок.

Потому что, у клиентов банка, которых нет детей, они тратят только на свои (а может и для супруги/супруга) основные социально-бытовые нужды, помимо самого погашения кредита, т.е. им не придётся пратить за алименты и расходы, связанные с нормальным жизнеобеспечением детей. 

##### 0.9.0.2 Вопрос 2:

Судя по таблице № 4, составленной по методу .pivot_table(), имеется неявная зависимость между семейным положением и возвратом кредита в срок.

Формулируется данным утверждением: большинство клиентов банка - это люди, которые имеют супружеские связи у данных клиентов банка и у них больше вероятность возрата кредита в срок. А неявная зависимость, потому что среди клиентов банка, которые уже имели супружеские связи, а теперь остались одни - скорее исключение, чем правило.  

Скорее всего, это связано с тем, что если клиент банка - многодетный семьянин, то ему нужно зарабатывать больше денег, чтобы обеспечивать супругу (или супруга) и детей (если они есть) (редкий позитивный случай: если он живёт на приусадебнем участке или у него есть дача (в СНТ, в селе или в деревне), то он может прокормить супругу (или супруга) на своём хозяйстве), то скорее он сможет заплатить за долги по кредитам, но если его неожиданно сократили на работе, то он может попасть в зоне риска должников.

##### 0.9.0.3 Вывод 3:

Судя по таблицам № 2 и 5, составленной по методу .pivot_table(), имеется неявная зависимость между уровнем дохода и возвратом кредита в срок. 

Формулируется следующим однозначным утверждением: чем выше уровень дохода у данных клиентов банка (не ниже уровня C, среди них большинство клиентов имеют категорию по уровню доходов 'С' и 'В'), тем больше вероятность возрата кредита в срок. А неявная зависимость, свидетельствует о том, что категории 'A', 'D', 'E' - это редкость, чем правило, потому что с доходами свыше миллиона рублей в месяц (категория 'A') могут похвастаться не многие, так как миллиона рублей в месяц могут зарабатывать только директорами, начальниками, топ-менеджерами в ведущих и иностранных компаниях, крупнейшими бизнесменами или депутатами в партийных номерклуатурах, а доходы в 50000 рублей в месяц и менее могут не потянуть на погашение кредитов, после оплаты социально-бытовых нужд и могут оказаться в должниках. 

Это очевидный факт, поскольку клиент всегда может расплатиться с банком по кредиту, в отличие от ситуации, когда он меньше получает, то он не может погасить долги, и если он сможет, то вынужден жить в коммуналке, на хлебе и воде, чтобы не оказаться должником.

##### 0.9.0.4 Вопрос 4:

Судя по таблицам №2 и 6, составленной по методу .pivot_table(), большинство клиентов банка обращаются за кредитами, с целью получения или другими оперциями с недвижимостью, у которых нет задолженности по возврату кредитов, а самая неохотная цель для получения кредита - это проведение свадьбы, у тех клиентов, которые уже имели (или имеют) задолженность (-и) по кредиту. 

Поскольку люди не могут прожить без еды, здоровья и жилья, если первыми двумя пунктами они могут себе по своим средствам, а сразу заплатить за недвижимость могут позволить себе только единицы населения, не точно только клиентов банка, когда сделали прелобработку данных, а на остальные цели сами себе заработать, накопить, выбрать наиболее выгодные предложения по различным услугам, на осуществление мечты и купить.<br>

А проведение свадьбы - это далеко не самая важная обязанность для человека, тем более, если он - должник, который берёт кредит на осуществление свадьбы, то он полный глупец в своих поступках, так как сэкономить на свадьбе - это единственная ценность, из четырёх перечисленных которую можно (и нужно) экономить без ущерба для клиента банка, в отличие от недвижимости (где нужно выбирать от расположения на территории, чтобы удобно ездить на работу до дополнительной инфраструктуры близкое расположение до станции метро, парка или магазинов), от автомобиля (чтобы не выбрать т.е. битую машину, где она побывала в ДТП, которую внешне и по документам пробразили, как в фразеологизме, "в конфетку"), получение образование (теоретически и практически, возможно сэкономить, если это не связано, с обучаением новой профессии или поступлением в вузы (реже ссузы) на договорной основе, которое не оплачивает компания, в которой этот клиент работает или когда его дети пойдут учится в вузы - получать хорошее высшее образование).
